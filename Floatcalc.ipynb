```python
import pandas as pd
df = pd.read_csv('/Users/aj/Documents/Texas State Docs/Fall 2025/Machine Learning/Project/ML-Project/Retrieval/Data_Combined/combined_matches_players_all_1.csv')
float_columns = df.select_dtypes(include=['float64', 'float32']).columns
num_float_columns = len(float_columns)
print(f"Number of columns with float data types: {num_float_columns}")
print(f"Float columns: {list(float_columns)}")
total_float_values = 0
for column in df.columns:
    total_float_values += df[column].apply(lambda x: isinstance(x, (float))).sum()
print(f"Total number of individual float values in the dataset: {total_float_values}")

num_entries = len(df)
print(f"Total number of entries (rows) in the dataset: {num_entries}")
```
